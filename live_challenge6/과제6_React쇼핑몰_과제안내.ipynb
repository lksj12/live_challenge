## 2. 과제 배경 시나리오

> 사용자는 상품을 조회하고 Firebase로 로그인하며, 원하는 상품을 전역 장바구니에 담아 예상 총액을 확인합니다.

### 인증과 장바구니 범위

- 로그인 기능은 필수이지만 회원 등급·주문 내역·프로필 관리는 제외합니다.
- 로그인 사용자별 장바구니 DB 저장은 필수가 아닙니다.
- 로그인 전 상품 열람·장바구니 허용 정책은 선택하고 README에 기록합니다.
- 인증 상태와 장바구니 상태의 소유 위치를 분리합니다.

### 사용자 흐름

```text
앱 실행
→ Firebase 인증 초기 상태
→ 상품 데이터 loading
→ 상품 목록 또는 오류·대체 데이터
→ 로그인 또는 사용자 상태
→ 상품 담기
→ 전역 장바구니·총액
→ 로그아웃
```

### 제외 범위

| 구분 | 기준 |
|---|---|
| 결제·카드·주문·배송 | 구현하지 않습니다. |
| 회원 등급·관리자 권한 | 구현하지 않습니다. |
| 전화번호 인증·다중 소셜 로그인 | 제외합니다. |
| 실제 개인정보 | README·캡처·mock에 포함하지 않습니다. |
| 사용자별 cart DB 저장 | 도전 설계만 허용합니다. |
| 결과 예시·브랜드·이미지 | 무단 복제하지 않습니다. |

## 3. 데이터·상태·저장 구조 설계

### TypeScript 노트 타입 설계

```ts
// 예시 구조이며 그대로 복사하는 정답 코드가 아닙니다.
interface Note {
  id: string;
  title: string;
  content: string;
  createdAt: string;
  updatedAt: string;
}
```

교육생은 프로젝트 구조에 맞게 타입 위치와 필드를 결정하고, `any` 사용을 최소화합니다.

### 노트 데이터 구조표
| 필드 후보 | 필요 | 자료형 후보 | 생성·수정 규칙 | 화면 사용 위치 |
|---|:---:|---|---|---|
| `id` |  | 문자열/숫자 | 중복 방지 | 선택·수정·삭제 |
| `title` |  | 문자열 | 공백 처리 정의 | 목록·편집 |
| `content` |  | 문자열 | 빈 내용 허용 정의 | 상세·편집 |
| `createdAt` |  | 날짜/문자열 | 생성 시 기록 | 표시 |
| `updatedAt` |  | 날짜/문자열 | 수정 시 갱신 | 정렬·표시 |

### 상태관리 구조 결정

| 구분 | 기본 권장 위치 | 예 |
|---|---|---|
| 노트 원본 데이터 | 상태관리 store | `notes` |
| 선택 노트 | store 또는 파생 상태 | `selectedNoteId` |
| 검색·정렬 조건 | store 또는 화면 공용 상태 | `query`, `sortBy` |
| 작성 폼 입력 | 컴포넌트 로컬 상태 가능 | `title`, `content` |
| 모달·메시지 | 컴포넌트 로컬 상태 가능 | `isOpen`, `message` |

동일한 노트 원본 데이터를 store와 컴포넌트 state에 중복 저장하지 않습니다.

### 기본 사용 제한
- 독립형 Antigravity의 다중 에이전트 병렬 운영은 기본 과제에서 사용하지 않습니다.
- `/goal`로 앱 전체 완성을 맡기지 않습니다.
- `/schedule`, 자동 배포, 원격 대량 작업을 사용하지 않습니다.
- MCP 서버와 플러그인은 기본 미사용이며 강사 승인 시에만 사용합니다.
- `Accept all`보다 `Review changes`에서 파일별 변경을 먼저 확인합니다.
- Implementation Plan 없이 바로 코딩하면 작업을 중단하고 계획 검토부터 다시 시작합니다.

### 명시적 승인 없이 수행 금지
